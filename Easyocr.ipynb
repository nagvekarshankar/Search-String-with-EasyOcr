```python
import easyocr

reader = easyocr.Reader(['en','hi'],gpu=False) # this needs to run only once to load the model into memory
```

```python
# show an image
import PIL
import pandas as pd
from PIL import ImageDraw
img= "detecting-sentiments-dataset/Sample Data Files/test.jpg"
img='paraimage.jpeg'
img='DLimage1.jpeg'
#img='hindi-english.jpg'
im = PIL.Image.open(img)
#im = PIL.Image.open("france.jfif")

im
```

```python
# Doing OCR. Get bounding boxes.
bounds = reader.readtext(img )
bounds
```

```python
type(bounds)
```

```python
data = pd.DataFrame(bounds, columns = ['Coordinates','Text','Confidence Score'])
data
```

```python
# Draw bounding boxes
def draw_boxes(image, bounds, color='purple', width=2):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image

draw_boxes(im, bounds)
```
